```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.datasets import fetch_california_housing
```

```python
from sklearn.metrics import mean_squared_error as mse, r2_score as r_2

# Follow this link to find more metrics for regression:
# https://scikit-learn.org/stable/modules/classes.html#sklearn-metrics-metrics
def regression_metrics(y_train, y_train_pred, y_test, y_test_pred):
    '''report the mse and r2 scores for both the training and test sets
    '''
    print('MSE Train: ', round(mse(y_train, y_train_pred), 3))
    print('MSE Test: ', round(mse(y_test, y_test_pred), 3))
    print('R^2 Train: ', round(r_2(y_train, y_train_pred), 3))
    print('R^2 Test: ', round(r_2(y_test, y_test_pred), 3))
```

#### Load the dataset

```python
#Load the California housing dataset
ds = fetch_california_housing()
X = ds.data
y = ds.target
print('dataset size:', X.shape, y.shape)
print('Data in ds', dir(ds))
print(ds.DESCR)
```

```python
# data split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
print('training set size:', X_train.shape[0])
print('test set size:', X_test.shape[0])
```

#### 1 Train and evaluate a linear regression model using OLS. 20 points.
    1) Train a linear reguression model
    2) Evalate the model and print out the RMS and $R^2$ values on the training and test sets
    3) What issues can you observe from the results? any possible solutions ?

```python
#OLS: ordinary least squares. 
from sklearn.linear_model import LinearRegression as LR

#1) train. 5 points
LR = LR()
LR.fit(X_train, y_train)

#2) evaluate. 5 points 
y_train_pred = LR.predict(X_train)
y_test_pred = LR.predict(X_test)

#3) report the mse and r2 scores for both the training and test sets. 
regression_metrics(y_train, y_train_pred, y_test, y_test_pred)
```

```python
# Decision tree
from sklearn.tree import DecisionTreeRegressor as DTR

#1) train. 10 points
DTR = DTR()
DTR.fit(X_train, y_train)

#2) pirnt out tree attributes: the depth of the tree, 
#the number of leaves, and the importance of each feature. 10 points
print('Depth of the tree: ', DTR.get_depth())
print('Number of leaves: ', DTR.get_n_leaves())
print('Importance of each feature: ', DTR.feature_importances_)

#3) evaluate. 5 points
DTR_y_train_pred = DTR.predict(X_train)
DTR_y_test_pred = DTR.predict(X_test)
regression_metrics(y_train, DTR_y_train_pred, y_test, DTR_y_test_pred)
```

```python
##### 6) Question: Test different max_depth values (5, 8, 10, 20) and analyse the results? 15 points.
#  - SET criterion = 'mse' and min_samples_leaf = 20
from sklearn.tree import DecisionTreeRegressor as DTR

tree1 = DTR(max_depth = 5, criterion = 'friedman_mse', min_samples_leaf = 20)

#  - Compare the results with different max_depth values
tree1.fit(X_train, y_train)
tree1_y_train_pred = tree1.predict(X_train)
tree1_y_test_pred = tree1.predict(X_test)
regression_metrics(y_train, tree1_y_train_pred, y_test, tree1_y_test_pred)
```

##### 1) What is the main problem of bagging approach? and how random forests can address the problem? 5 points

Response: The main disadvantage of the bagging approach is that it may introduce a loss of interpretability of the model. The resulting model can experience significant bias. 
Random forests can help address this problem by randomly selecting subsets of features used in each data sample. This wil help with the iterpretability-accuracy tradeoff. 

```python
# Random forests
from sklearn.ensemble import RandomForestRegressor as RFR

# 2) train. 10 points
RFR = RFR()
RFR.fit(X_train, y_train)

#3) evaluate. 5 points. 10 extra points for R^2>0.8
RFR_y_train_pred = RFR.predict(X_train)
RFR_y_test_pred = RFR.predict(X_test)
regression_metrics(y_train, RFR_y_train_pred, y_test, RFR_y_test_pred)
```
